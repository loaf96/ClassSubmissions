```python
import pandas as pd
import datetime as dt
```

```python
with open('nhtsa+complaints.csv', encoding='utf-8') as nhtsa_data:
    road_data = pd.read_csv(nhtsa_data)
road_data.head()
```

```python
# drop unnecessary columns
rd_data = pd.DataFrame(road_data[['Manufacturer', 'Crash', 'Fire', 'NumberOfInjured', 'DateofIncident_formatted', 'Make']])
rd_data.head()
```

```python
# drop all rows that do not have "Yes" as the response for whether or not there was a crash
data = {'Manufacturer':[], 'Crash':[], 'Fire':[], 'NumberOfInjured':[], 'DateOfIncident':[], 'Make':[]}
for x in rd_data.values:
    if 'Yes' in x:
        q = list(x)
        data['Manufacturer'].append(q[0])
        data['Crash'].append(q[1])
        data['Fire'].append(q[2])
        data['NumberOfInjured'].append(q[3])
        data['DateOfIncident'].append(q[4])
        data['Make'].append(q[5])
        next
    else:
        next
```

```python
# finalzed form of DF as dat_data with all rows dropped that did not have a Date for when the incident occured
dat_data = pd.DataFrame(data).dropna(how='any')
```

```python
with open('GM.csv') as gm_data:
    dat_gm = pd.read_csv(gm_data)
dat_gm.head()
```

```python
dat_var = dat_data['DateOfIncident'] > '2010-08-31'
dat_varr = dat_data[dat_var]
dat_varr.to_csv('MotorvehicleIncidents.csv')
```
